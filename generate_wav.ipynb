```python
from math import sin, cos, pi
import sympy

objs = []

# I
rect_points = [(0.0499, 0.4), (0.05, -0.4), (0.08, -0.4), (0.0801, 0.4)]
rect = sympy.Polygon(*rect_points)
objs.append(rect)

# heart
# triangle
p1, p2, p3 = map(sympy.Point, [(0.25, 0.3), (0.45, -0.4), (0.65, 0.3)])
rect = sympy.Polygon(p1, p2, p3)
objs.append(rect)
# circle
radius = 0.1
circle_res = 20
circle_points = [(0.35 + cos(2*pi/circle_res*x)*radius,0.3 + sin(2*pi/circle_res*x)*radius) for x in range(0,circle_res+1)]
poly = sympy.Polygon(*circle_points)
objs.append(poly)
# circle
radius = 0.1
circle_res = 20
circle_points = [(0.55 + cos(2*pi/circle_res*x)*radius,0.3 + sin(2*pi/circle_res*x)*radius) for x in range(0,circle_res+1)]
poly = sympy.Polygon(*circle_points)
objs.append(poly)

# U
circle_res = 20
radius = 0.1
circle_points = [(0.85 + cos(2*pi/circle_res*x)*radius,-0.4 + sin(2*pi/circle_res*x)*radius) for x in range(0,circle_res+1)][10:]
radius = 0.08
circle_points2 = [(0.85 + cos(2*pi/circle_res*x)*radius,-0.4 + sin(2*pi/circle_res*x)*radius) for x in range(0,circle_res+1)][10:]
circle_points2.reverse()
circle_points.extend(circle_points2)
poly = sympy.Polygon(*circle_points)
objs.append(poly)
rect_points = [(0.7499, 0.4), (0.75, -0.4), (0.77, -0.4), (0.7701, 0.4)]
rect = sympy.Polygon(*rect_points)
objs.append(rect)
rect_points = [(0.9301, 0.4), (0.93, -0.4), (0.95, -0.4), (0.9501, 0.4)]
rect = sympy.Polygon(*rect_points)
objs.append(rect)


# render data by tracing rays
res = 1000
upper = []
lower = []
for i in range(res):
    intersections = []
    for obj in objs:
        ray = sympy.Line(sympy.Point(i/res, 1.0), sympy.Point(i/res, -1.0))
        intersections.extend(obj.intersection(ray))

    if len(intersections) != 0:
        if type(intersections[0]) == sympy.Segment2D:
            p1 = intersections[0].p1.y.evalf(2)
            p2 = intersections[0].p2.y.evalf(2)
            if p1 > p2:
                upper.append(p1)
                lower.append(p2)
            else:
                upper.append(p2)
                lower.append(p1)
        else:
            upper.append(max(intersections, key=lambda p: p.y.evalf(2)).y.evalf(2))
            lower.append(min(intersections, key=lambda p: p.y.evalf(2)).y.evalf(2))
    else:
        upper.append(0)
        lower.append(0)

data = []
while len(upper) > 0 and len(lower) > 0:
    data.append(upper.pop(0))
    data.append(lower.pop(0))

with open("data", "w") as file:
    file.writelines([str(x) + "\n" for x in data])
```
